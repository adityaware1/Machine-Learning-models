```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/18_PCA/Exercise/heart.csv')
df
```

```python
df['Sex']=df.Sex.map({'M':1,'F':0})
```

```python
df
```

```python
dum=pd.get_dummies(df['ChestPainType'])
dum=dum.drop('TA',axis='columns')
dum
```

```python
df1=pd.concat([df,dum],axis='columns')
y=df1.HeartDisease
df1=df1.drop(['ChestPainType','HeartDisease'],axis='columns')
df1.ExerciseAngina=df1.ExerciseAngina.map({'N':0,'Y':1})
df1.ST_Slope=df1.ST_Slope.replace({'Up':3,'Flat':2,'Down':1})
df1.RestingECG.replace({'Normal':1,'ST':2,'LVH':3},inplace=True)
df1
```

```python
from sklearn.preprocessing import MinMaxScaler
ms= MinMaxScaler()
df1_ms=ms.fit_transform(df1)
```

```python
df1_ms
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df1_ms,y,test_size=0.2,random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

# using pca


```python
from sklearn.decomposition import PCA
pca=PCA(1)
pca_x=pca.fit_transform(df1)
```

```python
from sklearn.model_selection import train_test_split
x_ptrain,x_ptest,y_ptrain,y_ptest=train_test_split(pca_x,y,test_size=0.2,random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier()
model1.fit(x_ptrain,y_ptrain)
model1.score(x_ptest,y_ptest)
```
