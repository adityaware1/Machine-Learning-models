```python
from sklearn.datasets import load_digits
import pandas as pd
d=load_digits()
dir(d)
df=pd.DataFrame(d.data,columns=d.feature_names)
df
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
t={
    'SVM':
   {'model':SVC(gamma='auto'),
    'pm':{'C':[10,20,30],'kernel':['rbf','linear']}
   },
   
  'RandomForest':
  {'model':RandomForestClassifier(),
  'pm':{'n_estimators':[5,10,15]}
  },
    
  'logistic':
  {'model':LogisticRegression(),
  'pm':{'C': [1,5,10]}
  },
    
  'Gaussian':{
       
       'model':GaussianNB(),
       'pm':{}
   },
    
   'Multinomial':{
       
       'model':MultinomialNB(),
       'pm':{}
   },
    
   'tree':{
       'model':DecisionTreeClassifier(),
       'pm':{'criterion':['gini','entropy']}
   }
}



   
   
```

```python
score=[]
for k,v in t.items():
    
    clf=GridSearchCV(v['model'],v['pm'],return_train_score=False,cv=5) 
    clf.fit(df,d.target)
    score.append({'model':k,
                 'bestscore':clf.best_score_,
                 'bestparams':clf.best_params_})
    
    
```

```python
s=pd.DataFrame(score,columns=['model','bestscore','bestparams'])
s
```
